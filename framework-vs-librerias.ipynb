                                    📔 Frameworks y librerias
Estas dos herramientas resultan indispensables para cualquier desarrollador web y móvil ya que mientras creamos nuestro código a veces lidiamos con problemáticas similares y frecuentes, allí es donde nos salvan los frameworks y las librerías porque lo más probable es que ya tengan la solución a nuestro problema.

Pero ¿qué es un framework? y ¿qué es una librería?

frameworks-y-librerias

Framework
Como su nombre lo dice es un marco o entorno de trabajo que a su vez contiene librerías y otras herramientas que nos dan un amplio margen de resolución de problemas, pero no sólo eso, sino que otro de sus objetivos es el de unificar todos los componentes de un proyecto.

Ahora si buscas algo más técnico podemos decir que un framework es un conjunto de herramientas que nos facilitan el desarrollo de software; para poder lograrlo incluyen implementos como estos:

✔️APIs.
✔️Librerías.
✔️Herramientas de depuración.
✔️Edición.
✔️Prototipado.
✔️Programas de soporte.
✔️librerias-y-frameworks-

 

Librerías
Una librería es una sección de código que resuelve o brinda soporte para detalles específicos en nuestros programas o proyecto.

Ahora de forma más particular encontramos que las librerías son un conjunto de elementos, por ejemplo: funciones, clases, tipos predefinidos, constantes, variables globales y macros, etc.                                

Diferencias:
Conociendo bien los conceptos de cada uno es más sencillo poder distinguir sus diferencias. Una de estas es que el framework tiene un margen más amplio para resolver problemas ya que precisamente puede tener varias librerías, sin embargo, en el caso de estas últimas, aunque cuentan con otras herramientas solo sirven para determinadas ocasiones y los frameworks pueden ser usados en diferentes situaciones.

                        📑 Desarrollo front-end: frameworks vs. bibliotecas
El desarrollo front-end se refiere a la parte del desarrollo web que implica la creación de la interfaz y la experiencia de usuario (UI/UX) con la que los usuarios interactúan en sus navegadores. Se trata de lo que ven los usuarios y cómo interactúan con un sitio web.

                          📑Bibliotecas en desarrollo front-end
En el contexto del desarrollo front-end, las bibliotecas se utilizan normalmente para simplificar tareas específicas, como la manipulación del DOM, el control de animaciones o la administración del estado. Ofrecen herramientas útiles para los desarrolladores, pero no dictan cómo debe estructurarse su aplicación.

📝Ejemplos de bibliotecas front-end:

✅jQuery: jQuery es una de las bibliotecas de JavaScript más populares. Simplifica cosas como la manipulación del DOM, el manejo de eventos y las animaciones. Proporciona una API concisa y fácil de usar para interactuar con elementos HTML.

✅Axios: Axios es una biblioteca de JavaScript para realizar solicitudes HTTP. Simplifica la obtención de datos de las API y el manejo de operaciones asincrónicas con promesas.

✅Lodash: Lodash es una biblioteca de utilidades que proporciona funciones auxiliares para trabajar con matrices, objetos y otras estructuras de datos. Es especialmente útil para simplificar el código y hacerlo más legible.

Cómo funcionan las bibliotecas en el desarrollo front-end:

✔️Las bibliotecas son modulares y se centran en funcionalidades específicas.

✔️Los importa en el proyecto y utiliza los métodos u objetos proporcionados.

✔️Usted mantiene el control total sobre la estructura y el flujo de la aplicación.

                     Frameworks en el desarrollo front-end
Por otro lado, los marcos front-end proporcionan un enfoque más completo para crear aplicaciones. Proporcionan una estructura predefinida y, a menudo, incluyen soluciones integradas para cosas como el enrutamiento, el enlace de datos y las plantillas.

Ejemplos de frameworks front-end:

✅React.js: A menudo considerado una biblioteca, React es técnicamente un marco de front-end debido a sus convenciones arquitectónicas. React le permite crear interfaces de usuario interactivas utilizando componentes y administración de estado. React proporciona una estructura sobre cómo se deben organizar los componentes y cómo deben interactuar, lo que facilita la creación de aplicaciones de interfaz de usuario complejas.

✅Angular: Angular, desarrollado por Google, es un marco de front-end completo. Ofrece enlace de datos bidireccional, un potente sistema de plantillas e inyección de dependencias. Angular alienta a los desarrolladores a seguir un patrón arquitectónico estricto (MVC o MVVM) y lo ayuda a organizar su aplicación con una clara separación de preocupaciones.

✅Vue.js: Vue es un marco de JavaScript progresivo para crear interfaces de usuario. Es similar a React en muchos aspectos, pero proporciona soluciones más integradas, como un sistema de plantillas y un enlace de datos bidireccional. Vue es especialmente apreciado por su simplicidad y facilidad de integración en proyectos existentes.

Cómo funcionan los frameworks en el desarrollo front-end:

✔️Los marcos proporcionan una estructura de aplicación clara y hacen cumplir las convenciones de codificación.

✔️Incluyen funcionalidad integrada, por lo que no tiene que reinventar la rueda para cosas como enrutamiento, formularios o animaciones.

✔️El marco a menudo dicta cómo debe fluir y organizarse la aplicación.

                  Desarrollo de back-end: frameworks vs. bibliotecas
El desarrollo de back-end se centra en el lado del servidor de las aplicaciones web. Esto implica crear la lógica, las interacciones de la base de datos, las API y la configuración del servidor con los que los usuarios no interactúan directamente, pero que impulsan el front-end.

Bibliotecas en el desarrollo de back-end
Al igual que en el desarrollo front-end, las bibliotecas back-end ayudan a los desarrolladores al ofrecer funciones y herramientas preescritas para tareas comunes. Estas tareas pueden incluir el control de solicitudes HTTP, la interacción con bases de datos o la administración de la autenticación de usuarios. Las bibliotecas le ayudan a realizar el trabajo de manera eficiente, pero no dictan cómo se debe estructurar el resto de la aplicación.

Ejemplos de bibliotecas de back-end:

✅Express.js (para Node.js): Express es un marco de trabajo de aplicaciones web ligero para Node.js que ayuda con el enrutamiento, el manejo de solicitudes HTTP y la creación de API. Si bien técnicamente es un marco, a menudo se considera una biblioteca porque se enfoca en simplificar el proceso de configuración de un servidor y el manejo de solicitudes.

✅Sequelize: Sequelize es una biblioteca ORM (Object-Relational Mapping) para Node.js que le permite interactuar con bases de datos SQL de forma orientada a objetos. Abstrae las consultas de la base de datos en funciones de JavaScript y proporciona funciones eficaces como migraciones y validación de modelos.

✅Passport.js: Passport es una biblioteca para la autenticación de usuarios en aplicaciones Node.js. Simplifica la integración de varias estrategias de autenticación como OAuth, Facebook, Google y autenticación local.

Cómo funcionan las bibliotecas en el desarrollo de back-end:

Las bibliotecas se centran en funcionalidades específicas, como las consultas a la base de datos o la autenticación de usuarios.

Puede incorporarlos a su aplicación sin aplicar una arquitectura específica.
Usted decide dónde y cómo usar la biblioteca dentro de su base de código.

Frameworks en el desarrollo de back-end
Los marcos de back-end, por otro lado, son más completos. Proporcionan una estructura completa para el desarrollo de aplicaciones web, incluidos aspectos como el enrutamiento, la gestión de bases de datos, la autenticación de usuarios y el middleware. Los marcos ofrecen soluciones integradas para problemas comunes de back-end y aplican las mejores prácticas para mantener un código limpio y escalable.

Ejemplos de marcos de back-end:

✅Node.js con Express: Si bien Express puede verse como una biblioteca, a menudo se usa como parte del marco de back-end Node.js. Node.js le permite escribir JavaScript para el lado del servidor, mientras que Express simplifica el enrutamiento y el manejo de solicitudes HTTP.

✅Django (para Python): Django es un marco web de Python de alto nivel que fomenta el desarrollo rápido y el diseño limpio y pragmático. Incluye herramientas integradas para cosas como la autenticación de usuarios, las migraciones de bases de datos y las plantillas, todas siguiendo una estructura específica.

✅Ruby on Rails: Rails es un sólido marco de back-end para Ruby que utiliza el patrón arquitectónico Model-View-Controller (MVC). Viene con herramientas integradas para manejar el enrutamiento, la administración de bases de datos y la representación de vistas, lo que agiliza el proceso de desarrollo.

✅Spring Boot (para Java): Spring Boot es un marco que simplifica la creación de aplicaciones web basadas en Java. Proporciona un vasto ecosistema de herramientas para el acceso a datos, la seguridad, la mensajería y más, todo integrado en una estructura coherente.

Cómo funcionan los frameworks en el desarrollo de back-end:

Los frameworks proporcionan una estructura completa y un conjunto de convenciones para crear aplicaciones web.
A menudo vienen con herramientas integradas para el enrutamiento, la interacción con la base de datos, la autenticación y más.

El marco dicta el flujo general de la aplicación y aplica ciertos patrones, lo que lleva a un código más fácil de mantener y escalable.
